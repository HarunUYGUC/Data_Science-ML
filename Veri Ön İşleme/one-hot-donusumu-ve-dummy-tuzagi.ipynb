### One-Hot Dönüşümü ve Dummy Değişken Tuzağı

```python
import seaborn as sns
df = sns.load_dataset('tips')
df.head()
```

```python
df.head()
```

```python
import pandas as pd
```

```python
df_one_hot = pd.get_dummies(df, columns = ["sex"], prefix = ["sex"])
```

```python
df_one_hot.head()
```

```python
pd.get_dummies(df, columns = ["day"], prefix = ["day"]).head()
```
