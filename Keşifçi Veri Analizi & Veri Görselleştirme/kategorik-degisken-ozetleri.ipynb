## Kategorik Değişken Özetleri

```python
import seaborn as sns 
planets = sns.load_dataset("planets")
df = planets.copy()
df.head()
```

### Sadece Kategorik Değişkenler ve Özetleri

```python
kat_df = df.select_dtypes(include = ["object"])
```
